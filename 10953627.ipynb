## Null Count
---
_Null count_ refers to the number of missing values in a particular column in a DataFrame.
To determine the number of missing values in each column, we use:

```python
null_counts = df.isnull().sum()
```

```python
print(null_counts)
```

There are ten null count values.

To determine the location of the null values:

```python
# Find the locations of null values in this DataFrame
import numpy as np
null_locations = np.where(df.isnull())

# Print the locations of null values
for row, col in zip(null_locations[0], null_locations[1]):
    print(f"Null value at row {row} and column {df.columns[col]}")
```

## Duplicate Values
---
The duplicated() method returns a series with True and False values that describe which row in the DataFrame is duplicated and not


To determine the number of duplicates and their locations:

```python
# Find the number of duplicates
duplicates=df.duplicated()
print(duplicates)
number_duplicates = df.duplicated().sum()


# Print the number of duplicates
print(f"There are {number_duplicates} duplicate rows in the DataFrame.")

# Find the duplicate rows
duplicate_rows = df[df.duplicated()]

# Print the duplicate rows
print("Duplicate rows in the DataFrame:")
print(duplicate_rows)
```

## Pre-processing DataFrame
---
To pre-process data to get rid of all null values and duplicates:

```python
# Print the original DataFrame
print('Original DataFrame:\n',df)

# Remove null count values
# Remove any rows with missing values
df = df.dropna()

# Print the DataFrame without null count values
print('DataFrame without null count values:\n', df)
```

```python
 #Print the original DataFrame
print('Original DataFrame:\n',df)


# Remove duplicate rows
df = df.drop_duplicates()

# Print the DataFrame without duplicates
print('DataFrame without duplicates:\n',df)
```

This analysis shows there are no duplicates. 

## Using the Dropna method for Null values
---
The reason for using the Dropna module is that the 'dropna()' method in Pandas allows to remove missing values from a DataFrame by either dropping rows or columns that contain any missing values or only those that contain a certain threshold of missing values.This can be beneficial to provide you with only valid data. This method is verysimple, effective, flexible and can be customized to fit specific needs.

Hidden inferences in Jupyter Dataframe refer to the truncated rows and columns in large datasets to make them more manageable for viewing. The default format is displaying the first and last five rows and hiding those in the middle. Hidden inferences help to make data analysis and manipulation more efficient and user-friendly.Some hidden inferences in Jupyter are; column, index, missing values, operations etc.

```python
 import seaborn as sns
histogram_plot1 = sns.histplot(data=df, x='fixed acidity')

histogram_plot1.set_title("Histogram plot for fixed acidity")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
mean_1 = df['fixed acidity'].mean()


above_mean_count = df[df['fixed acidity'] > mean_1]['fixed acidity'].count()
below_mean_count = df[df['fixed acidity'] < mean_1]['fixed acidity'].count()


labels = ['Greater than Mean', 'Less than Mean']
sizes = [above_mean_count, below_mean_count]
colors = ['red', 'cyan']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%' )
# plt.axis('equal')
plt.title('Fixed Acidity Greater than and Less than Mean')
plt.show()
```
